赤しその発現データとisoformのデータをローカスデータとjoinしてtsvに書き出します

注
- Enzymeをキーにデータフレームでmergeしている
- F3PHにローカスのデータが無いためローカルなxref_idがつかずPathwayダイアグラムでノードをクリックしても何も起こらない
- xref_idがNaNの場合0を与えているが、同じようなケースが今後もあると思うのでXref IDの値として０を振るのは避けたほうが良いかも

```python
import csv
import pandas as pd
```

```python
file_1 = "/Users/oec/Desktop/data/202401_広島大学_赤しそ/SuppTbl4Fig3.tsv"
file_2 = "/Users/oec/Desktop/data/202401_広島大学_赤しそ/PP.tsv"
```

```python
df1 = pd.read_csv(file_1, delimiter="\t")
df1
```

```python
df2 = pd.read_csv(file_2, delimiter="\t")
df2
```

```python
df = pd.merge(df1, df2, on="Enzyme", how="left")
pd.set_option("display.max_rows", None)
df
```

```python
# DF2のEnzymeが紐づかない行はxref_idにNaNが入る。xref_idはintにキャストするのだがその際NaNはエラーになるためデフォルト値として０を与える

df["xref_id"].fillna(0, inplace=True)
df["xref_id"] = df["xref_id"].astype("int")
df
```

```python
df.to_csv("red_perilla_anthocyanin_test.tsv", sep="\t", index=False)
```
